#özellik mühendisliği

```python
#özellik mühendisliği
df['Date']
```

```python
df['Date'].dtype #tarih olmasına rağmen veri türü object
```

```python
#veri türünü değiştirir.
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].info()
```

```python
df['Day']=df.Date.dt.day  # Date sütunundaki günü al ve day sütun oluşturup
df['Day']
```

```python
#ay ı ayır.
df['Month']=df.Date.dt.month
df['Month']
```

```python
df['year']=df.Date.dt.year
df['year']
```

```python
df['dayofweek']=df.Date.dt.dayofweek
df['dayofweek']
```

```python
df['dayofweek'].unique()
```

```python
#haftanın kaçıncı günü olduğu yerine gün adı yazsın
df['dayofweek'].replace({0:'Monday',1: 'Thuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6: 'Sunday'},inplace=True)
```

```python
df['dayofweek']
```

```python
#ikinci yöntem
df['day_name']=df.Date.dt.day_name()
df['day_name']
```

```python
df['month_name']=df.Date.dt.month_name()
df['month_name']
```

```python
df['Time'],df['Time'].dtype
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Time'].info()
```

```python
df['hour']=df.Time.dt.hour
df['hour']
```

```python
plt.figure(figsize=(12,6)) # grafiğin boyutları
sns.countplot(x=df['month_name']) 
# belirlenen kategorilerdeki toplamları alır
#Aylardaki alışveriş sayısı
plt.title('Aylardaki alışveriş sayısı')
plt.xticks(rotation=90) # etiket yazısını açısı
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Şueberin aylara göre alışveriş sayısı')
sns.countplot(x=df['month_name'],
              hue=df['Branch'])
```

```python
plt.title('Cinsiyete göre alışveriş sayısı')
sns.countplot(x=df['Gender'])
```

```python
plt.title('Cinsiyete göre aylık alış veriiş sayıları')
sns.countplot(x=df['month_name'], hue=df['Gender'])
```
